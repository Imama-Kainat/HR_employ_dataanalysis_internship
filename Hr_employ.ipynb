```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/HR-Employee-Attrition.csv')
print(df)
```

```python
# Data Cleaning:

# Deleting Redundant Columns:
df = df.drop(columns=['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'])
print(df)
```

```python
# Renaming Columns:
df = df.rename(columns={'Attrition': 'Attrition_Status', 'JobInvolvement': 'Job_Involvement'})
print(df)
```

```python
# Dropping Duplicates:
df = df.drop_duplicates()
print(df)
```

```python
# Cleaning Individual Columns:
df['BusinessTravel'] = df['BusinessTravel'].str.strip()
df['MaritalStatus'] = df['MaritalStatus'].str.strip()
df['OverTime'] = df['OverTime'].str.strip()
print(df)
```

```python
df = df.dropna()
print(df)
#  removes rows with missing values (NaN) from  df
```

```python
# Data Visualization:

# Plotting a Correlation Map:
# correlation map for all numeric variables in the DataFrame df
fig, ax = plt.subplots(figsize=(12, 12))  # Set the figure size
cax = ax.matshow(df.corr(), cmap='coolwarm')  # Use the 'coolwarm' colormap for better readability
plt.xticks(range(df.shape[1]), df.columns, rotation=90)
plt.yticks(range(df.shape[1]), df.columns)
plt.colorbar(cax)  # Use the same colorbar as the correlation map
plt.show()


```

```python
# Visualizing Attrition by Education Field:
df.groupby('EducationField')['Attrition_Status'].value_counts().unstack().plot(kind='bar', stacked=True)
plt.xlabel('Education Field')
plt.ylabel('Attrition Count')
plt.title('Attrition by Education Field')
plt.show()
# a stacked bar chart showing the count of attrition for each education field.
```

```python
# Visualizing Attrition by Overtime:
df['OverTime'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Attrition by Overtime')
plt.ylabel('')
plt.show()
#  plots a pie chart showing the percentage of attrition for employees who work overtime versus those who do not.
```

```python
# Create a box plot for Age grouped by Attrition_Status
sns.boxplot(x='Attrition_Status', y='Age', data=df)
plt.title('Box Plot of Age by Attrition Status')
plt.xlabel('Attrition Status')
plt.ylabel('Age')
plt.show()
#  a box plot showing the distribution of ages for employees with and without attrition.
```
